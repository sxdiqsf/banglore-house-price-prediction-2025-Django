# Data-Preprocessing

```python
df.head()
```

```python
df.columns
```

```python
df_clean = df.copy()
df_clean.head()
```

```python
df_clean['location'] = df_clean['location'].map({'BTM Layout': 0, 'Banashankari': 1, 'Bannerghatta Road': 2, 'Bellandur': 3, 'Electronic City': 4, 'HSR Layout': 5, 'Hebbal': 6, 'Indiranagar': 7, 'JP Nagar': 8, 'Jayanagar': 9, 'KR Puram': 10, 'Kengeri': 11, 'Koramangala': 12, 'Malleshwaram': 13, 'Marathahalli': 14, 'Rajajinagar': 15, 'Sarjapur Road': 16, 'Uttarahalli': 17, 'Whitefield': 18, 'Yelahanka': 19})

df_clean['furnishing'] = df_clean['furnishing'].map({'Unfurnished': 0, 'Semi-Furnished': 1, 'Fully-Furnished': 2})

df_clean['property_type'] = df_clean['property_type'].map({'Apartment': 0, 'Independent House': 1, 'Villa': 2})
```

```python
df_clean['price_per_sqft'] = df_clean['price'] / df['area']
df_clean['bhk_per_bath'] = df_clean['bhk'] / df['bath']
```

```python
df_clean.head()
```

# Features and Target

```python
x = df_clean.drop('price',axis = 1)
y = df_clean['price']
```

# Train Test and Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(x_train)
X_test_sc = scaler.transform(x_test)
```

# using linear regresssion

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Train model
lr = LinearRegression()
lr.fit(X_train_sc, y_train)

# Predictions
y_pred_lr = lr.predict(X_test_sc)

# Evaluation
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("ðŸ”¹ Linear Regression Results:")
print(f"MAE: {mae_lr:,.2f}")
print(f"RMSE: {rmse_lr:,.2f}")
print(f"RÂ² Score: {r2_lr:.4f}")
```

```python
def predict_house_price_lr(area, location, bhk, bath, balcony, parking, furnishing, property_type, age):
    # Create DataFrame for input
    new_data = pd.DataFrame({
        'area': [area],
        'location': [location],          # numeric id (0â€“19)
        'bhk': [bhk],
        'bath': [bath],
        'balcony': [balcony],
        'parking': [parking],
        'furnishing': [furnishing],      # 0â€“2
        'property_type': [property_type],# 0â€“2
        'age': [age]
    })

    # Derived feature
    new_data['bhk_per_bath'] = new_data['bhk'] / new_data['bath']

    # Compute average price_per_sqft for this location
    avg_pps = df_clean[df_clean['location'] == location]['price_per_sqft'].mean()

    if np.isnan(avg_pps):
        avg_pps = df_clean['price_per_sqft'].mean()  # fallback to global avg
    
    new_data['price_per_sqft'] = avg_pps

    # Reorder columns to match training
    new_data = new_data[df_clean.drop('price', axis=1).columns]

    # Apply same scaler
    new_data_scaled = scaler.transform(new_data)

    # Predict
    predicted_price = lr.predict(new_data_scaled)[0]

    return f"ðŸ  Predicted Price using linear regression: â‚¹{predicted_price:,.0f}"


print(predict_house_price_lr(
    area=1800,
    location=18,          # Whitefield
    bhk=3,
    bath=2,
    balcony=1,
    parking=1,
    furnishing=1,         # Semi-Furnished
    property_type=0,      # Apartment
    age=5
))
```

# using randomforestregressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print("\nðŸŒ³ Random Forest Results:")
print(f"MAE: {mae_rf:,.2f}")
print(f"RMSE: {rmse_rf:,.2f}")
print(f"RÂ² Score: {r2_rf:.4f}")
```
